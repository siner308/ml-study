```python
train_df['Age'].fillna(train_df['Age'].dropna().median(), inplace=True)
test_df['Age'].fillna(test_df['Age'].dropna().median(), inplace=True)

for dataset in combine:
    dataset.loc[dataset['Age'] <= 8, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 8) & (dataset['Age'] <= 16), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 24), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 24) & (dataset['Age'] <= 32), 'Age'] = 3
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 40), 'Age'] = 4
    dataset.loc[(dataset['Age'] > 40) & (dataset['Age'] <= 48), 'Age'] = 5
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 56), 'Age'] = 6
    dataset.loc[(dataset['Age'] > 56) & (dataset['Age'] <= 64), 'Age'] = 7
    dataset.loc[(dataset['Age'] > 64) & (dataset['Age'] <= 72), 'Age'] = 8
    dataset.loc[(dataset['Age'] > 72) & (dataset['Age'] <= 80), 'Age'] = 9
    dataset.loc[dataset['Age'] > 80, 'Age'] = 10
    
    dataset['Age'] = dataset['Age'].astype(int)
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 20)
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
train_df['Fare'].fillna(train_df['Fare'].dropna().median(), inplace=True)
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)

for dataset in combine:
    dataset.loc[dataset['Fare'] <= 10, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 10) & (dataset['Fare'] <= 20), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 20) & (dataset['Fare'] <= 30), 'Fare'] = 2
    dataset.loc[(dataset['Fare'] > 30) & (dataset['Fare'] <= 40), 'Fare'] = 3
    dataset.loc[(dataset['Fare'] > 40) & (dataset['Fare'] <= 50), 'Fare'] = 4
    dataset.loc[(dataset['Fare'] > 50) & (dataset['Fare'] <= 70), 'Fare'] = 5
    dataset.loc[(dataset['Fare'] > 70) & (dataset['Fare'] <= 90), 'Fare'] = 6
    dataset.loc[(dataset['Fare'] > 90) & (dataset['Fare'] <= 100), 'Fare'] = 7
    dataset.loc[dataset['Fare'] > 100, 'Fare'] = 8
    dataset['Fare'] = dataset['Fare'].astype(int)
    
train_df.head()
```

```python
for dataset in combine:
    dataset.loc[dataset['PassengerId'] <= 50, 'PassengerId'] = 0
    dataset.loc[(dataset['PassengerId'] > 50) & (dataset['PassengerId'] <= 100), 'PassengerId'] = 1
    dataset.loc[(dataset['PassengerId'] > 100) & (dataset['PassengerId'] <= 150), 'PassengerId'] = 2
    dataset.loc[(dataset['PassengerId'] > 150) & (dataset['PassengerId'] <= 200), 'PassengerId'] = 3
    dataset.loc[(dataset['PassengerId'] > 200) & (dataset['PassengerId'] <= 300), 'PassengerId'] = 4
    dataset.loc[(dataset['PassengerId'] > 300) & (dataset['PassengerId'] <= 400), 'PassengerId'] = 5
    dataset.loc[(dataset['PassengerId'] > 400) & (dataset['PassengerId'] <= 500), 'PassengerId'] = 6
    dataset.loc[(dataset['PassengerId'] > 500) & (dataset['PassengerId'] <= 600), 'PassengerId'] = 7
    dataset.loc[(dataset['PassengerId'] > 600) & (dataset['PassengerId'] <= 700), 'PassengerId'] = 8
    dataset.loc[(dataset['PassengerId'] > 700) & (dataset['PassengerId'] <= 800), 'PassengerId'] = 9
    dataset.loc[(dataset['PassengerId'] > 800) & (dataset['PassengerId'] <= 900), 'PassengerId'] = 10
    dataset.loc[dataset['PassengerId'] > 900, 'PassengerId'] = 11

combine = [train_df, test_df]
train_df.head()
```

```python
X_train = train_df.drop('Survived', axis=1)
Y_train = train_df['Survived']
X_test = test_df.copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
# Support Vector Machines

svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```
